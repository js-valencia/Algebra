```python
import numpy as np

vector1 = np.array([-2, 5])
vector2 = np.array([3, 4])

sum_vector = vector1 + vector2

print(sum_vector)
```

```python
scalar = 1/2
scaled_vector = scalar * vector1

print(scaled_vector)
```

```python
scalar2 = -4
scaled_vector2 = scalar2 * vector2

print(scaled_vector2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the origin for the vectors
origin = [0], [0]

# Create the plot
plt.figure(figsize=(10, 8))

# Plot original vectors
plt.quiver(*origin, vector1[0], vector1[1], color='r', scale=1, scale_units='xy', angles='xy', label='Vector 1')
plt.quiver(*origin, vector2[0], vector2[1], color='g', scale=1, scale_units='xy', angles='xy', label='Vector 2')

# Plot sum vector
plt.quiver(*origin, sum_vector[0], sum_vector[1], color='b', scale=1, scale_units='xy', angles='xy', label='Sum Vector (V1 + V2)')

# Plot scaled vectors
plt.quiver(*origin, scaled_vector[0], scaled_vector[1], color='purple', scale=1, scale_units='xy', angles='xy', label='Scaled Vector 1 (0.5 * V1)')
plt.quiver(*origin, scaled_vector2[0], scaled_vector2[1], color='orange', scale=1, scale_units='xy', angles='xy', label='Scaled Vector 2 (-4 * V2)')

# Determine max absolute value for setting plot limits
all_vectors = np.array([vector1, vector2, sum_vector, scaled_vector, scaled_vector2])
max_val = abs(all_vectors).max() + 2 # Add some padding

# Set plot limits and labels
plt.xlim(-max_val, max_val)
plt.ylim(-max_val, max_val)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Combined Vector Operations Visualization')
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box') # Ensure equal aspect ratio
plt.legend()
plt.show()
```
