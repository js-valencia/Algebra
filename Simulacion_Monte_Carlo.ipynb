```python
import random

n = 10000  # número de lanzamientos

# Inicializar contador
resultados = [0, 0, 0, 0, 0, 0]

# Simulación
for i in range(n):
    dado = random.randint(1, 6)
    resultados[dado - 1] += 1

# Mostrar frecuencias
print("-----------------------------------")
print("Resultados de los lanzamientos:")
for i in range(6):
    print(f"Cara {i+1}: {resultados[i]} veces")

# Probabilidades
print("-----------------------------------")
print("\nProbabilidades:")
for i in range(6):
    prob = resultados[i] / n
    print(f"Cara {i+1}: {prob:.4f}")

# Comparación con teórica
print("-----------------------------------")
print("\nComparación con probabilidad teórica (1/6 ≈ 0.1667):")
for i in range(6):
    prob = resultados[i] / n
    error = abs(prob - (1/6))
    print(f"Cara {i+1}: {prob:.4f} | Error: {error:.4f}")
```

```python
import matplotlib.pyplot as plt

# Gráfica
caras = [1, 2, 3, 4, 5, 6]
probabilidades = [r/n for r in resultados]

plt.bar(caras, probabilidades)
plt.axhline(y=1/6, linestyle='--')
plt.xlabel("Caras del dado")
plt.ylabel("Probabilidad")
plt.title("Simulación Monte Carlo - Lanzamiento de dado")
plt.show()
```

Se realizó una simulación de 10,000 lanzamientos de un dado utilizando el método de Monte Carlo en Python. Se generaron números aleatorios entre 1 y 6 para representar cada lanzamiento. Los resultados obtenidos muestran que las frecuencias de cada cara tienden a ser similares, aproximándose a la probabilidad teórica de 1/6. Esto confirma la Ley de los Grandes Números, donde al aumentar el número de simulaciones, los resultados se acercan al valor esperado.
